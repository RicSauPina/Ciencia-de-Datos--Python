## Extrae en línea la base de datos “Iris.csv” de la siguiente liga y agrégala a un DataFrame en Python. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings

warnings.filterwarnings("ignore")

os.getcwd()

from sklearn.cluster import KMeans

IRIS = pd.read_csv("Iris.csv")
IRIS.head()
```

## Agrupa las observaciones de la tabla descargada mediante el algoritmo de K-Medias sin realizar ningún tipo de transformación de los datos originales. Para dicho efecto, determina el número óptimo de clusters tanto por el criterio del gráfico de codo así como por el indicador Silhouette.

```python
P_Sepal = IRIS.iloc[:, [0,1]].values
P_Petal = IRIS.iloc[:, [2,3]].values
```

### SEPAL

```python
# Codo Sepal
WCSS_P_S = []
for index in range(1,11):
    kmeans_S = KMeans(n_clusters = index, init = "k-means++", random_state= 21)
    kmeans_S.fit(P_Sepal)

    WCSS_P_S.append(kmeans_S.inertia_)

# Graficacipon del "Codo de Jambu"
sns.set()
plt.plot(range(1,11), WCSS_P_S)
plt.title("K-Means Clustering")
plt.xlabel("Numero de clusters")
plt.ylabel("WCSS") #Weighted cluster sum of squares
```

```python
#Criterio de Silueta ("Silhouette")
from sklearn import datasets
from sklearn.metrics import silhouette_score

P_Sepal_NP = IRIS.iloc[:, [0,1]].to_numpy()
for j in range(2,12):
    kmeans = KMeans(n_clusters= j, random_state= 21)
    kmeans.fit_predict(P_Sepal_NP)
    # Calculamos el Score de Silueta

    score = silhouette_score(P_Sepal_NP, kmeans.labels_, metric = "euclidean")
    print("Score Silhouette:", "k = ", j, ":", score)
```

```python
kmeans_S = KMeans(n_clusters=2, init="k-means++", random_state = 42)
cluster_values_S = kmeans_S.fit_predict(P_Sepal)
print(cluster_values_S)
```

### PETAL

```python
# Codo Petal
WCSS_P_P = []
for index in range(1,11):
    kmeans_P = KMeans(n_clusters = index, init = "k-means++", random_state= 21)
    kmeans_P.fit(P_Petal)

    WCSS_P_P.append(kmeans_P.inertia_)

# Graficacipon del "Codo de Jambu"
sns.set()
plt.plot(range(1,11), WCSS_P_P)
plt.title("K-Means Clustering")
plt.xlabel("Numero de clusters")
plt.ylabel("WCSS") #Weighted cluster sum of squares
```

```python
#Criterio de Silueta ("Silhouette")
from sklearn import datasets
from sklearn.metrics import silhouette_score

P_Petal_NP = IRIS.iloc[:, [2,3]].to_numpy()
for j in range(2,12):
    kmeans = KMeans(n_clusters= j, random_state= 21)
    kmeans.fit_predict(P_Petal_NP)
    # Calculamos el Score de Silueta

    score = silhouette_score(P_Petal_NP, kmeans.labels_, metric = "euclidean")
    print("Score Silhouette:", "k = ", j, ":", score)
```

```python
kmeans_P = KMeans(n_clusters=2, init="k-means++", random_state = 42)
cluster_values_P = kmeans_P.fit_predict(P_Petal)
print(cluster_values_P)
```

```python
plt.scatter(P_Sepal[cluster_values_S == 0, 0], P_Sepal[cluster_values_S == 0,1], c = "red", label = "Cluster 1")
plt.scatter(P_Sepal[cluster_values_S == 1, 0], P_Sepal[cluster_values_S == 1,1], c = "orange", label = "Cluster 2")
# plt.scatter(P_Sepal[cluster_values_S == 2, 0], P_Sepal[cluster_values_S == 2,1], c = "pink", label = "Cluster 3")
plt.scatter(kmeans_S.cluster_centers_[:,0], kmeans_S.cluster_centers_[:,1], c = "black", label = "Centroide")

plt.title("Grupos por Sepal")
plt.xlabel("length")
plt.ylabel("width")
```

```python
plt.scatter(P_Petal[cluster_values_P == 0, 0], P_Petal[cluster_values_P == 0,1], c = "red", label = "Cluster 1")
plt.scatter(P_Petal[cluster_values_P == 1, 0], P_Petal[cluster_values_P == 1,1], c = "orange", label = "Cluster 2")
# plt.scatter(P_Petal[cluster_values_P == 2, 0], P_Petal[cluster_values_P == 2,1], c = "pink", label = "Cluster 3")
plt.scatter(kmeans_P.cluster_centers_[:,0], kmeans_P.cluster_centers_[:,1], c = "black", label = "Centroide")

plt.title("Grupos por Petal")
plt.xlabel("length")
plt.ylabel("width")
```

## Repite el ejercicio anterior después de haber transformado tus datos originales a dos columnas, mediante reducción por PCA. Gráfica los grupos obtenidos considerando las dos dimensiones resultantes.

```python
# Algoritmo KMeans aplicado a 4 columnas de nuestra base original
# Bajo una transformacón de PCA

IRIS_PS = IRIS.iloc[:, [0,1, 2, 3]].values
IRIS_PS
```

```python
from sklearn import decomposition
PCA = decomposition.PCA(n_components = 2)
PCA.fit(IRIS_PS)
IRIS_PS = PCA.transform(IRIS_PS)
IRIS_PS
```

```python
WCSS2 = []
for index in range(1,11):
    kmeans = KMeans(n_clusters = index, init = "k-means++", random_state= 23)
    kmeans.fit(IRIS_PS)

    WCSS2.append(kmeans.inertia_)

# Graficacipon del "Codo de Jambu"
sns.set()

plt.plot(range(1,11), WCSS2)
plt.title("K-Means Clustering")
plt.xlabel("Numero de clusters")
plt.ylabel("WCSS") #Weighted cluster sum of squares
```

```python
#Criterio de Silueta ("Silhouette")
from sklearn import datasets
from sklearn.metrics import silhouette_score

IRIS_PS_NP = IRIS.iloc[:, [0,1, 2, 3]].to_numpy()
for j in range(2,12):
    kmeans = KMeans(n_clusters= j, random_state= 21)
    kmeans.fit_predict(IRIS_PS_NP)
    # Calculamos el Score de Silueta

    score = silhouette_score(IRIS_PS_NP, kmeans.labels_, metric = "euclidean")
    print("Score Silhouette:", "k = ", j, ":", score)
```

```python
kmeans_PS = KMeans(n_clusters=2, init="k-means++", random_state = 42)
cluster_values = kmeans_PS.fit_predict(IRIS_PS)
print(cluster_values)
```

```python
plt.scatter(IRIS_PS[cluster_values == 0, 0], IRIS_PS[cluster_values == 0,1], c = "red", label = "Cluster 1")
plt.scatter(IRIS_PS[cluster_values == 1, 0], IRIS_PS[cluster_values == 1,1], c = "orange", label = "Cluster 2")

plt.scatter(kmeans_PS.cluster_centers_[:,0], kmeans_PS.cluster_centers_[:,1], c = "black", label = "Centroide")

plt.title("Grupos de flores")
plt.xlabel("length")
plt.ylabel("width")

# mp.legend()
```
