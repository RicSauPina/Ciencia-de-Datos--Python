```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# os.getcwd()
```

```python
iris1 = pd.read_csv("Iris.csv") #Cargar base de datos de Iris
iris1
```

```python
etiquetas = iris1["variety"]
etiquetas
```

```python
from sklearn.decomposition import TruncatedSVD
A = iris1

#Remover columna no numerica
A = A.drop(["variety"], axis = 1)
A
```

```python
from numpy import diag
from numpy import zeros
from scipy.linalg import svd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
U, s, VT = svd(A) #Obtencion de matrices cuyo producto arroja la matriz A
Sigma = zeros((A.shape[0], A.shape[1])) #Creacion de matriz de 0 de la misma dimensión que A
Sigma[:A.shape[1], :A.shape[1]] = diag(s)
```

```python
print("Matriz U:")
print(U)
```

```python
print("Matriz Sigma:")
print(Sigma)
```

```python
print("Matriz VT:")
print(VT)
```

```python
# Para reconstruir la matriz original
B = U.dot(Sigma.dot(VT))
print(B)
```

```python
# Reduccion a c DOS Dimensiones
n_elements = 2

UReduced2 = U[:,:n_elements]
print("U reducida:")
print(UReduced2)
```

```python
SigmaReduced2 = Sigma[:n_elements, :n_elements]
print("Matriz Sigma Reducida")
print(SigmaReduced2)
```

```python
VTReduced2 = VT[:n_elements, :]
print("VT Reducida")
print(VTReduced2)
```

```python
AReduced2 = UReduced2.dot(SigmaReduced2.dot(VTReduced2))
print("Matriz A transformas:")
print(AReduced2)
```

```python
T2 = UReduced2.dot(SigmaReduced2)
print("Matriz T (2 Dimensiones):")
print(T2)
```

```python
df = pd.DataFrame(T2, columns= ["a","b"])
df.plot(kind= "scatter", x = "a", y = "b")
plt.show
```

```python
# Reduccion a 3 dimensiones
n_elements = 3

UReduced3 = U[:,:n_elements]

SigmaReduced3 = Sigma[:n_elements, :n_elements]

VTReduced3 = VT[:n_elements, :]

AReduced3 = UReduced3.dot(SigmaReduced3.dot(VTReduced3))
print("Matriz A transformas:")
print(AReduced3)
```

```python
T3 = UReduced3.dot(SigmaReduced3)
print("Matriz T (2 Dimensiones):")
print(T3)
```
